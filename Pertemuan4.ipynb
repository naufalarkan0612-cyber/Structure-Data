Queue dengan List

```python
# Python program to demonstrate queue implementation using list

queue = []

# append() function to enqueue element in the queue
queue.append('a')
queue.append('b')
queue.append('c')
queue.append('d')
queue.append('e')

print('Initial queue')
print(queue)

# pop() function to dequeue element from queue in FIFO order
print('\nElements dequeued from queue:')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nQueue after removing elements')
print(queue)

# uncommenting print(queue.pop(0)) will cause an IndexError as the queue is now empty
```

Stack dengan Collections.deque

```python
# Python program to demonstrate queue implementation using collections.deque
from collections import deque

queue = deque()

# append() function to Enqueue element in the queue
queue.append('a')
queue.append('b')
queue.append('c')
queue.append('d')
queue.append('e')

print("Initial queue:")
print(queue)

# popleft() function to dequeue element from queue in FIFO order
print("\nElements dequeued from the queue")
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print("\nQueue after removing elements")
print(queue)

# uncommenting print(queue.popleft()) will cause an IndexError as the queue is now empty
```

Queue dengan queue.Queue

```python
# Python program to demonstrate queue implementation using queue module
from queue import Queue

# Initializing a queue
qu = Queue(maxsize=5)

# qsize() show the number of elements in the queue
print(qu.qsize())

# put() function to enqueue element in the queue
qu.put('a')
qu.put('b')
qu.put('c')
qu.put('d')
qu.put('e')

print("Full: ", qu.full())
print("Size: ", qu.qsize())

# get() function to dequeue element from queue
print("\nElements dequeued from the queue")
print(qu.get())
print(qu.get())
print(qu.get())

print("\nEmpty: ", qu.empty())
print("Full: ", qu.full())
```

Queue dengan Single Linked List

```python
class Queue:
    def __init__(self):
        self.queue = list()

    def addtoquq(self, dataval):
        # Insert method to add element
        if dataval not in self.queue:
            self.queue.insert(0, dataval)
            return True
        return False

    # Pop method to remove element
    def removefromq(self):
        if len(self.queue) > 0:
            return self.queue.pop()
        return ("No elements in Queue!")

    def size(self):
        return len(self.queue)


TheQueue = Queue()
print(TheQueue.addtoquq("Jan"))
print(TheQueue.addtoquq("Feb"))
print(TheQueue.addtoquq("March"))
print(TheQueue.addtoquq("Aprsl"))
print(TheQueue.size())
print(TheQueue.removefromq())
print(TheQueue.removefromq())
print(TheQueue.size())
```
